Mise  jour de la classe stock :

```python
import java.util.ArrayList;

public class Stock {
    private static Stock instance;
    private ArrayList<String> items;


    private Stock() {
        items = new ArrayList<String>();
    }

    // public method to provide access to the instance
    public static Stock getInstance() {
        if (instance == null) {
            instance = new Stock();
        }
        return instance;
    }

    public void addItem(String item) {
        items.add(item);
    }

}
```

Mise  jour de la classe EditorGui :

```python
import javax.swing.*;
import java.awt.event.ActionEvent;
import java.awt.event.ActionListener;

public class EditorGui extends JFrame implements ActionListener {
    private JTextField itemField;
    private JButton addButton;
    private JTextArea displayArea;

    public EditorGui() {
        setTitle("Stock Editor");
        setSize(400, 300);
        setDefaultCloseOperation(JFrame.EXIT_ON_CLOSE);
        setLocationRelativeTo(null);

        itemField = new JTextField(20);
        addButton = new JButton("Add Item");
        displayArea = new JTextArea();
        displayArea.setEditable(false);

        JPanel panel = new JPanel();
        panel.add(new JLabel("Item:"));
        panel.add(itemField);
        panel.add(addButton);

        add(panel, "North");
        add(new JScrollPane(displayArea), "Center");

        addButton.addActionListener(new ActionListener() {
            @Override
            public void actionPerformed(ActionEvent e) {
                String item = itemField.getText();
                if (!item.isEmpty()) {
                    Stock.getInstance().addItem(item);
                    displayArea.append(item + "\n");
                    itemField.setText("");
                }
            }
        });
    }

    public static void main(String[] args) {
        SwingUtilities.invokeLater(new Runnable() {
            @Override
            public void run() {
                new EditorGui().setVisible(true);
            }
        });
    }
}
```
